# 1. Theory <a class="anchor" id="1.Theory"></a>

### Stationarity

A stationary time series is one whose properties do not depend on the time at
which the series is observed. Thus, time series with trends, or with
seasonality, are not stationary — the trend and seasonality will a|ect the value
of the time series at di|erent times. On the other hand, a white noise series is
stationary — it does not matter when you observe it, it should look much the
same at any point in time. 

In order to make predicitve models, a time series needs to be transformed into a stationary form.
Following charts demonstrates methods which can be implemented to test stationarity of a time series.
![Alt text](attch/Stationarity.png?raw=true "Title")

### Elements of Time Series
Data points are often non-stationary or have means, variances and covariances that change over time. Non-stationary behaviors can be **Trends, Cycles, Random Walks** or combinations of the three.

Non-stationary data, as a rule, are unpredictable and cannot be modeled or forecasted. The results obtained by using non-stationary time series may be spurious in that they may indicate a relationship between two variables where one does not exist. In order to receive consistent, reliable results, the non-stationary data needs to be transformed into stationary data. In contrast to the non-stationary process that has a variable variance and a mean that does not remain near, or returns to a long-run mean over time, the stationary process reverts around a constant long-term mean and has a constant variance independent of time.

### Residuals(Random Walk) of Time Series

The “residuals” in a time series model are what is left over after fitting a model.
For many (but not all) time series models, the residuals are equal to the
difference between the observations and the corresponding fitted values.

Residuals are useful in checking whether a model has adequately captured the
information in the data. A good forecasting method will yield residuals with the
following properties:

**1. The residuals are uncorrelated. If there are correlations between residuals,
then there is information left in the residuals which should be used in
computing forecasts.**

**2. The residuals have zero mean. If the residuals have a mean other than zero,
then the forecasts are biased.**

Any forecasting method that does not satisfy these properties can be improved.
However, that does not mean that forecasting methods that satisfy these
properties cannot be improved. It is possible to have several di|erent
forecasting methods for the same data set, all of which satisfy these properties.
Checking these properties is important in order to see whether a method is
using all of the available information, but it is not a good way to select a
forecasting method.

In addition to these essential properties, it is useful (but not necessary) for the
residuals to also have the following two properties.

**3. The residuals have constant variance.**

**4. The residuals are normally distributed.**


### Models: Time Series

**1. Autoregressive Models - AR(p)**<br>
The notation $AR(p)$ indicates an autoregressive model of order $p$. The AR(p) model is defined as

\begin{equation*}
X_t = c+ \sum_{i=1}^p \phi_i X_{i-1} + \epsilon_t
\end{equation*}

where $\phi_1 ,..., \phi_p$ are the parameters of the model,  $c$ is a constant, and $\epsilon_t$ is white noise. 
___
**2. Moving Average Models - MA(q)**<br>
The notation $MA(q)$ refers to the moving average model of order $q$:

\begin{equation*}
X_t = \mu + \sum_{i=1}^q \theta_i \epsilon_{t-i} + \epsilon_t
\end{equation*}

where $μ$ is the mean of the series, the $\theta_1, ..., \theta_q$ are the parameters of the model and the $\epsilon_t, ..., \epsilon_{t-q}$ are white noise error terms. 
___
**3. Autoregressive Moving Average Models - ARMA(p, q)**<br>
The notation $ARMA(p, q)$ refers to the model with $p$ autoregressive terms and $q$ moving-average terms. This model contains the $AR(p)$ and $MA(q)$ models.

\begin{equation*}
X_t = c+ \sum_{i=1}^p \phi_i X_{i-1} + \sum_{i=1}^q \theta_i \epsilon_{t-i} +  \epsilon_t
\end{equation*}
___
**4. Autoregressive Integrated Moving Average Models - ARIMA(p', q)**

Given a time series of data $X_t$ where $t$ is an integer index and the $X_t$ are real numbers, an $ARMA(p',q)$ model is given by

\begin{equation*}
\left(1-\sum_{i=1}^{p'} \alpha_i L_{i}\right)X_t = \left(1+\sum_{i=1}^{q} \theta_i L_{i}\right)\epsilon_t 
\end{equation*}

\begin{equation*}
L^k X_t = X_{t-k}
\end{equation*}

where $L$ is the lag operator, the $\alpha_i$ are the parameters of the autoregressive part of the model, the  $\theta_i$ are the parameters of the moving average part and the $\epsilon_t$ are error terms. The error terms $\epsilon_t$ are generally assumed to be independent, identically distributed variables sampled from a normal distribution with zero mean.
___
### Models: Variance of Time Series


**A. Autoregressive Conditionally Heteroskedastic Models - ARCH(p)**<br>
In econometrics, the autoregressive conditional heteroskedasticity (ARCH) model is a statistical model for time series data that describes the variance of the current error term  as a function of the actual sizes of the previous time periods' error terms; often the variance is related to the squares of the previous innovations. The ARCH model is appropriate when the error variance in a time series follows an autoregressive (AR) model.

To model a time series using an ARCH process, let $\epsilon_t$ denote the error terms (return residuals, with respect to a mean process), i.e. the series terms. These $\epsilon_t$ are split into a stochastic piece $z_t$ and a time-dependent standard deviation $\sigma_t$ characterizing the typical size of the terms so that

\begin{equation*}
\epsilon_t = \sigma_t z_t
\end{equation*}

The random variable $z_t$ is a strong white noise process. The series $\sigma_t^2$ is modeled by

\begin{equation*}
\sigma_t^2 = \alpha_0 + \sum_{i=1}^q \alpha_i \epsilon_{t-i}^2
\end{equation*}

where $\alpha_0>0~and~\displaystyle \alpha_i\geq 0,~i>0.$
___
**B. Generalized Autoregressive Conditionally Heteroskedastic Models - GARCH(p,q)**<br>
If an autoregressive moving average model (ARMA) model is assumed for the error variance, the model is a generalized autoregressive conditional heteroskedasticity (GARCH) model.

In that case, the $GARCH (p, q)$ model (where $p$ is the order of the GARCH terms $\sigma ^2$ and $q$ is the order of the $ARCH$ terms $\epsilon^2$ is given by

\begin{equation*}
{\displaystyle y_{t}=x'_{t}b+\epsilon _{t}}
\end{equation*}

\begin{equation*}
{\displaystyle \epsilon _{t}|\psi _{t-1}\sim {\mathcal {N}}(0,\sigma _{t}^{2})}
\end{equation*}

\begin{equation*}
{\displaystyle \sigma _{t}^{2}=\omega +\alpha _{1}\epsilon _{t-1}^{2}+\cdots +\alpha _{q}\epsilon _{t-q}^{2}+\beta _{1}\sigma _{t-1}^{2}+\cdots +\beta _{p}\sigma _{t-p}^{2}=\omega +\sum _{i=1}^{q}\alpha _{i}\epsilon _{t-i}^{2}+\sum _{i=1}^{p}\beta _{i}\sigma _{t-i}^{2}} 
\end{equation*}

### Helper Functions

```python
def tsplot(y, dtindex=[]):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    if len(dtindex)>0:
        y.index = dtindex
    y.dropna(inplace=True)
    
    fig = plt.figure(figsize=(15, 10))

    layout = (3, 2)
    ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
    acf_ax = plt.subplot2grid(layout, (1, 0))
    qq_ax = plt.subplot2grid(layout, (1, 1))
    hist_ax = plt.subplot2grid(layout, (2, 0), colspan=2)

    y.plot(ax=ts_ax)
    ts_ax.set_title('Time Series Analysis Plots')
    ts_ax.grid(True)
    autocorrelation_plot(y,ax=acf_ax)
    acf_ax.grid(True)
    sm.qqplot(y, line='s', ax=qq_ax)
    qq_ax.grid(True)
    qq_ax.set_title('QQ Plot') 
    y_sorted = y.sort_values()
    mu = y_sorted.mean()
    sigma = y_sorted.std()
    hist_ax.set_title(f"Probability Distribution Plot - N({round(mu,3)}, {round(sigma,3)})")
    y.hist(ax=hist_ax,bins=int(len(y)/15), density=True)
    hist_ax.set_ylabel("Frequency")
    hist_ax = plt.plot(y_sorted, scs.norm.pdf(y_sorted, mu, sigma), 'r')

    plt.tight_layout()
    return 
```

ARIMA model of statsmodels covers all first 4 models described in previous section.
When using arima the method of maximizing likelihood comes into play, which is defined as following:<br>

**method: {‘css-mle’,’mle’,’css’}**<br>
This is the loglikelihood to maximize. If “css-mle”, the conditional sum of squares likelihood is maximized and its values are used as starting values for the computation of the exact likelihood via the Kalman filter. If “mle”, the exact likelihood is maximized via the Kalman Filter. If “css” the conditional sum of squares likelihood is maximized. All three methods use start_params as starting parameters.

**trend {‘c’,’nc’}**<br>
Whether to include a constant or not. ‘c’ includes constant, ‘nc’ no constant.

## 2.1 Predictive models

#### AR, MA, ARMA, ARIMA
Using arima method in statsmodel, it's possible to model all predictive methods mentioned above.
For each model, the following input to arima is used.

**AR**: q=d=0<br>
**MA**: p=d=0<br>
**ARMA**: d=0 and p,q != 0<br>
**ARIMA**: p,d,q != 0

```python
%%time

P = Q = [0,1,2,3,4]
D = [0,1]
arima = {"Model":[],
         "Type":[],
         "Instance":[],
        "Log Likelihood":[],
         "AIC":[],
         "BIC":[],
        "HQIC":[],
        "MSE":[]}

for p in P:
    for q in Q:
        for d in D:
            try:
                mdl = smt.ARIMA(df0_2.values, order=(p,d,q)).fit(method='mle', trend='nc')
                arima["Model"].append(f"ARIMA({p},{d},{q})")
                
                if q==0 and d==0:
                    _type = "AR"
                elif p==0 and d==0:
                    _type = "MA"
                elif d==0 and p!=0 and q!=0:
                    _type = "ARMA"
                elif d!=0 and p!=0 and q!=0:
                    _type = "ARIMA"
                else:
                    _type = "_"
                
                arima["Type"].append(_type)
                arima["Instance"].append(mdl)
                arima["Log Likelihood"].append(mdl.llf)
                arima["AIC"].append(mdl.aic)
                arima["BIC"].append(mdl.bic)
                arima["HQIC"].append(mdl.hqic)
                arima["MSE"].append(sum(map(lambda x: x ** 2, mdl.resid))/len(mdl.resid))
                
            except:
                continue
```

```python
df_results = pd.DataFrame(arima).sort_values("Log Likelihood")
df_results = df_results[df_results.Type!="_"].reset_index(drop=True)

df_results
```

The generated table depicts, the performace of the all 4 models. Following conclusion can be inferred:

    1. ARMA has the best and ARIMA has the least performance. Given that one differentiating had been implemented, it's expected a new differentiation would be unnecessary. So ARIMA model which is not needed.
    
    2. 5 performance metrics have been used with Likelihood being the main one.

```python
ax = df_results[["AIC", "MSE"]].plot(figsize=(15,5), xticks=df_results.index, secondary_y="MSE", grid=True)
ax = ax.set_xticklabels(df_results.Model, rotation='vertical')
```

It can be seen that ARIMA(4,0,2) has the best performance.
